```python
with open('../output/calc.txt') as file:
    lines = file.readlines()
    arr = list(map(lambda s: list(map(float, s[:-3].split(', '))), lines))
    plt.figure(figsize=(10, 7))
    plt.pcolormesh(arr)
    plt.colorbar()
```
